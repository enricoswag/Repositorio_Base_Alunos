# funçao Print ()
a Funçao print eh uma das mais importantes e usadas na linguagem Phyton. sua funçao é, basicamente, exibir mensagens na tela ou envia-las para outro dispositivo. como imprimir dentro de um arquivo novo.
No Phyton 3 print () é uma funçao interna, de um modo que nao é necessario importar nenhuma biblioteca para poder utiliza-la. basta chaa-la e passar os argumentos necessarios                                            

```python
print ("meu primeiro print!") # todo texto deve estar entre aspas simples ou duplas
```

```python
print("meu segundo print")
```

```python
print ("D'agua") # quando uma palavra contém aspas inicias devem ser diferentes
```

```python
print (20+15) # e o Phyton realiza a operação que esta dentro do parenteses
```

```python
print ("20+15") # A operação entre aspas sera apenas impressa nao executada, pois o phyton entende isso como texto
```

```python
print("20+15=" , 20+15)
```

```python
n1 = "1"
n2 = "3"
print (n1 + n2) # neste exemplo o Phyyton faz a concatenaçao de 2 strings
```

```python
nome = "alessa"
sobrenome = "cadeira voadora"
nome_completo = nome + " " + sobrenome
print (nome_completo)
```
