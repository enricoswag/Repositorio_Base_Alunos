## Hora Da Pratica

1) crie um codigo que receba o nome do aluno e informe a media do aluno
obs: no seu print voce deve apresentar o nome do aluno, as duas notas e a média

```python
nome_aluno = input("Digite seu nome: ")
nota1 = float(input("Digite a primeira nota 1: "))
n2 = float(input("Digite a segunda nota 2: "))
media = (nota1 + n2)/2
print(f"O aluno(a) {nome_aluno} teve notas {nota1}, {n2} e sua média {media}")
```

2. Crie um codigo que receba as medidas de um retangulo e informe o perimetro do mesmo

```python
tamanho1 = float(input("Digite a medida do retangulo(largura): "))
tamanho2 = float(input("Digite a medida do retangulo(altura): "))
perimetro = (tamanho1 + tamanho2)*2
print(f"O perimetro do retangulo é {perimetro}")
```

3. crie um codigo que calcule a raiz quadrada de um numero

```python
num1 = float(input("Digite um numero:"))
raiz = num1 ** 0.5
print(f"A raiz quadrada de  {num1} é igual a {raiz}.")
```

## 4. crie uma calculadora simples(adiçao, subtração,multiplicaçao,divisao.

```python
num1 = float(input("Digite o primeiro numero 1: "))
num2 = float(input("Digite o primeiro numero 2: "))
soma = num1 + num2
subtracao = num1 - num2
multiplicacao = num1 * num2
divisao = num1 / num2
print(f"A soma, multiplicacao, subtracao, divisao de {num1} + {num2} = {soma}, {multiplicacao}, {subtracao}, {divisao}")
```

## 5) calcule a diferença entre dois numeros

```python
numero=float(input("Digite o primeiro numero: "))
numero2=float(input("Digite outro numero: "))
subtração=numero-numero2
print(f"A diferença entre {numero} e {numero2} é {subtração}")
```

## 6) calcule o produto entre dois numeros

```python
num1 = float(input("Digite o primeiro numero 1: "))
num2 = float(input("Digite o primeiro numero 2: "))
produto = num1 * num2
print(f"O produto de {num1} * {num2} = {produto}")
```

## 7) crie um codigo que receba uma base e depois o expoente e calcule a potencia

```python
num=float(input("Digite um numero (base): "))
expoente=float(input("Digite o expoente: "))
potencia=num**expoente
print(f"A potencia de {num} elevado a {expoente} é {potencia}")
```
